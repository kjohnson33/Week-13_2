```python
import numpy as np

class Particle(object):
    
    def __intit__(self, mass, x,y,z, vx, vy, vz):
        self.mass = mass
        self.x = x 
        self.y = y
        self.z = z
        self.vx = vx
        self.vy = vy
        self.vz = vz
        
# Ques: 2        
    def ke_eq(self):
        return (0.5*self.mass)*(np.sqrt(((self.vx)**2) + ((self.vy)**2) + ((self.vz)**2)))**2
    
# Ques: 3
    def dist(self, other):
        return (np.sqrt(((self.x - other.x)**2) + ((self.y - other.y)**2) + ((self.z - other.z)**2)))
    
# Ques: 4 
    def time_new_posit(self, time):
        self.x = self.x + (self.vx*time)
        self.y = self.y + (self.vy*time)
        self.z = self.z + (self.vz*time)
```

```python
# write new code here

import numpy as np

class ChargedParticle(Particle):

    def __init__(self,mass, x, y, z, vx, vy, vz, charge):
        self.mass = mass
        self.x = x
        self.y = y
        self.z = z
        self.vx = vx
        self.vy = vy
        self.vz = vz
        self.charge = charge
        
    def charge(self, other):
        sumchar = self.charge + other.charge
        if (sumchar == 0):
            ps = (np.sqrt(((self.x - other.x)**2) + ((self.y - other.y)**2) + ((self.z - other.z)**2)))
            if (ps<0.25): 
                print("Interaction")
            self.vx = 0
            self.vy = 0
            self.vz = 0
            other.vx = 0
            other.vy = 0
            other.vz = 0
            self.charge = 0
            other.charge = 0
            print("Merge")

        else:  
            self.vx = self.vx * (-1*(sumchar)) * (self.mass/other.mass)
            self.vy = self.vy * (-1*(sumchar)) * (self.mass/other.mass)
            self.vz = self.vz * (-1*(sumchar)) * (self.mass/other.mass)
            other.vx = other.vx * (-1*(sumchar)) * (other.mass/self.mass)
            other.vy = other.vy * (-1*(sumchar)) * (other.mass/self.mass)
            other.vz = other.vz * (-1*(sumchar)) * (other.mass/self.mass)
            print("Repel")
```
